<h3>Task 3: Frequency Distribution Tables</h3>
<p>Creating frequency distribution tables and plotting their graphs</p>

```python
#frequency distribution table of all columns
for column in df.columns:
    freq_table = df[column].value_counts().reset_index()
    freq_table.columns = [column, 'frequency']
    print(freq_table)
    print('\n')
```

<p>Plotting frequency distribution graphs</p>

```python
#frequency distribution graphs for all columns
sns.set(font_scale=1.5)
def freq_dist(df):
    columns = ['Satisfaction_Level', 'Interaction_Challenge']
    for col in columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            col_data = df[col].dropna()
            if not col_data.empty:
                g = np.histogram(col_data, bins=5, density=False)
                col1 = []
                col2 = []
                for i in range(len(g[0])):
                    col1.append(g[0][i])
                    col2.append(f'{round(g[1][i], 2)} - {round(g[1][i+1], 2)}')
                freq_table = pd.DataFrame({f'{col}_distribution': col2, 'freq': col1})
                freq_table.set_index(f'{col}_distribution', inplace=True)
                display(freq_table)
                sns.histplot(col_data, kde=True, bins=8).set(title=col)
                plt.show()
                print("----------------------------------------------------------------------------------------------------")

# Call the function with your DataFrame
freq_dist(df)
```

<h3> Task 6 : Hypothesis Testing </h3>
<p>Taking different hypothesis and proving or rejecting them using hypothesis testing. </p>

<h4>Hypothesis 1</h4>

<p>Null Hypothesis (H0): There is no significant association between the level of satisfaction with educational experience and the level of institutional support during the pandemic.

Alternative Hypothesis (H1): There is a significant association between the level of satisfaction with educational experience and the level of institutional support during the pandemic.</p>

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Assuming df contains your data
# Create a contingency table
contingency_table = pd.crosstab(df['Satisfaction_Level'], df['Institutional_Support'])

# Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
```

<h5>Conclusion : </h5>

<p>p > 0.05, Fail to reject null hypothesis. There is insufficient evidence to claim a significant association.</p>

<h4>Hypothesis 2</h4>

<p>Null Hypothesis (H0): There is no significant difference in reported impact on mental health between remote, hybrid, and in-person learning environments.

Alternative Hypothesis (H1): There is a significant difference in reported impact on mental health between remote, hybrid, and in-person learning environments.</p>

```python
mapping = {'Yes': 1, 'No': 0, 'Maybe': 0.5}
df['Mental_Health'] = df['Mental_Health'].replace(mapping)
```

```python
from itertools import combinations
from scipy.stats import ttest_ind

groups = ['Remote', 'Hybrid', 'In-Person']
combs = list(combinations(groups, 2))  # Generate all pairwise combinations

for comb in combs:
    group1 = df[df['Learning_Environment'] == comb[0]]['Mental_Health']
    group2 = df[df['Learning_Environment'] == comb[1]]['Mental_Health']
    
    t_stat, p_val = ttest_ind(group1, group2, equal_var=False)  # Assume unequal variances
    
    print(f"T-test between {comb[0]} and {comb[1]}:")
    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_val}")
    print()

```

<h5>Conclusion : </h5>
<p>since the p-value for remote and hybrid environment is close to 0.05, we can carefully accept the hypothesis that there is significant difference in reported impact on mental health between remote and hybrid learning environments.</p>

<h4>Hypothesis 3</h4>

<p>Null Hypothesis (H0): There is no significant difference in the mean satisfaction levels between students in remote and hybrid learning environments.

Alternative Hypothesis (H1): There is a significant difference in the mean satisfaction levels between students in remote and hybrid learning environments.</p>

```python
from scipy.stats import ttest_ind

# Separate the satisfaction levels for remote and hybrid learning environments
satisfaction_remote = df[df['Learning_Environment'] == 'Remote']['Satisfaction_Level']
satisfaction_hybrid = df[df['Learning_Environment'] == 'Hybrid']['Satisfaction_Level']

# Perform the independent t-test
t_statistic, p_value = ttest_ind(satisfaction_remote, satisfaction_hybrid, equal_var=False)

# Check the p-value
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: There is a significant difference in satisfaction levels.")
else:
    print("Fail to reject null hypothesis: There is no significant difference in satisfaction levels.")
```

<h5>Conclusion</h5>
<p>since p > 0.05, we fail to reject the null hypothesis and there is no significant evidence for our claim that there is a  significant difference in the mean satisfaction levels between students in remote and hybrid learning environments. </p>
